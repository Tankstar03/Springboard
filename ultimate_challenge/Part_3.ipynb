## Data Description

● **city:** city this user signed up in\
● **phone:** primary device for this user\
● **signup_date:** date of account registration; in the form ‘YYYYMMDD’\
● **last_trip_date:** the last time this user completed a trip; in the form ‘YYYYMMDD’\
● **avg_dist:** the average distance in miles per trip taken in the first 30 days after signup\
● **avg_rating_by_driver:** the rider’s average rating over all of their trips\
● **avg_rating_of_driver:** the rider’s average rating of their drivers over all of their trips\
● **surge_pct:** the percent of trips taken with surge multiplier > 1\
● **avg_surge:** The average surge multiplier over all of this user’s trips\
● **trips_in_first_30_days:** the number of trips this user took in the first 30 days after signing up\
● **ultimate_black_user:** TRUE if the user took an Ultimate Black in their first 30 days; FALSE otherwise\
● **weekday_pct:** the percent of the user’s trips occurring during a weekday

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# read the json file containing the data
df = pd.read_json('ultimate_data_challenge.json')
```

```python
df.sort_values('last_trip_date', ascending=False)
```

```python
# The latest date is 2014-07-01, thus this was when the data was pulled.
# The cutoff date would then be 2014-06-01, marking the 30-day period.
# Create a new bool column determining if a user is retained or not
df['retained'] = np.where(df['last_trip_date'] < '2014-06-01', 0, 1)
df.head(30)
```

```python
df.isna().sum()
```

```python
# The null values are in columns of less importance in this context
df.drop(columns=['avg_rating_by_driver', 'phone', 'avg_rating_of_driver'], inplace=True)
```

```python
df.head(30)
```

```python
# All individuals above registered during January, want to predict if they are still active by July (retained)
# Look for any correlations with a heatmap
temp = df.drop(columns=['city'])
#pd.to_numeric(temp['signup_date'])
#pd.to_numeric(temp['last_trip_date'])
f, ax = plt.subplots(figsize=(12,10))
sns.heatmap(df.drop(columns=['city', 'signup_date', 'last_trip_date']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
f.suptitle("Heatmap for User Registration/Activity Data")
```

```python
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, learning_curve
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.dummy import DummyClassifier
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, accuracy_score, confusion_matrix
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

## Preprocessing

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='retained'), 
                                                    df.retained, test_size=0.3, 
                                                    random_state=30)
```

```python
X_train
```

```python
print(X_train.shape, X_test.shape)
```

```python
dummy = DummyClassifier(strategy='prior')
dummy.fit(X_train, y_train)
dummy.predict(X_test)
dummy.score(X_test, y_test)
```

```python
X_train = X_train.drop(columns=['city', 'signup_date', 'last_trip_date'])
X_test = X_test.drop(columns=['city', 'signup_date', 'last_trip_date'])
```

# Modeling

### Logistic Regression

```python
# Begin by building a logistic regression pipeline
pipe_lr = make_pipeline(
    SimpleImputer(strategy='mean'),
    StandardScaler(),
    LogisticRegression()
)
```

```python
pipe_lr.fit(X_train, y_train)
```

```python
y_pred_tr = pipe_lr.predict(X_train)
y_pred_te = pipe_lr.predict(X_test)
```

```python
r2 = r2_score(y_train, y_pred_tr), r2_score(y_test, y_pred_te)
r2
```

```python
RMSE = mean_squared_error(y_train, y_pred_tr, squared=False), mean_squared_error(y_test, y_pred_te, squared=False)
RMSE
```

```python
cv_scores = cross_validate(pipe_lr, X_train, y_train, cv=5)
cv_scores
```

```python
accuracy = accuracy_score(y_train, y_pred_tr), accuracy_score(y_test, y_pred_te)
accuracy
```

```python
cm_a = confusion_matrix(y_train, y_pred_tr)
```

```python
sns.heatmap(cm_a, 
            annot=True,
            fmt='g',
            xticklabels=['Not Retained','Retained'],
            yticklabels=['Not Retained','Retained'])
plt.title('Confusion Matrix (Training Data)', fontsize=17, pad=20)
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.show()
```

```python
cm_b = confusion_matrix(y_test, y_pred_te)
```

```python
sns.heatmap(cm_b, 
            annot=True,
            fmt='g',
            xticklabels=['Not Retained','Retained'],
            yticklabels=['Not Retained','Retained'])
plt.title('Confusion Matrix (Testing Data)', fontsize=17, pad=20)
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.show()
```

Compared to the dummy classifier, the logistic regressor performed better but not by much, 62.7% vs 68.5%. Let's see if we can do better.

After conducting Principle Component Analysis on the data, both scaled and not, seems like more important factors to pay attention to are whether or not an individual is an ultimate black user, and what their surge percentage is. Unscaled, the "elbow" is after the 2nd component, and scaled it is after the 6th component.

## Conclusion

The catboost model has proven to provide the most accurate results in predicting Ultimate's rider retention rate, at over 76%. Details to look for in determining whether a user is more likely to be retained would be their surge percent and whether or not they are an ultimate black user. The higher their surge percent and being an ultimate black user increases their chance of continuing to actively use their account. Assuming surge increases after every ride and decreases slowly over time, as well as ultimate black being the equivalent of uber black, perhaps ways to encourage greater ride usage would be to adjust the surge rate, optimize it so more users are compelled to be more active, and maybe an offer where the first ultimate black a user rides will be at a discounted rate.
