# Part 2 - Experiment and metrics design

The neighboring cities of Gotham and Metropolis have complementary circadian rhythms: on
weekdays, Ultimate Gotham is most active at night, and Ultimate Metropolis is most active
during the day. On weekends, there is reasonable activity in both cities.
However, a toll bridge, with a two way toll, between the two cities causes driver partners to tend
to be exclusive to each city. The Ultimate managers of city operations for the two cities have
proposed an experiment to encourage driver partners to be available in both cities, by
reimbursing all toll costs.
1) What would you choose as the key measure of success of this experiment in
encouraging driver partners to serve both cities, and why would you choose this metric?
2) Describe a practical experiment you would design to compare the effectiveness of the
proposed change in relation to the key measure of success. Please provide details on:
    1) how you will implement the experiment
    2) what statistical test(s) you will conduct to verify the significance of the
observation
    3) how you would interpret the results and provide recommendations to the city
operations team along with any caveats.

## Measurement for Success

The best metric would be the activity levels in each of the city, particularly those during the weekdays. In most cases, city managers would want to pursue such an experiment to help boost GDP and revenue. For simplicity, let's assume that city activity and GDP have a strong positive correlation to each other. The mark to beat would be whether or not the positive change in activity in both cities is greater than that generated by the tolls on the bridge. By reimbursing all toll costs, the tolls would no longer generate any revenue. Thus, if the GDP increase in both cities from the experiment is greater than the revenue lost from the elimination of the tolls, then the experiment can be considered a success.

### Experiment Setup

Depending on how urgent the cities are on boosting activity, designate a period of time, let's say 3 months, where all toll costs will be reimbursed. Each city will then over the next 3 month period monitor the daily activity of their own cities, with particular emphasis during the weekdays. Once the experimental phase is over (after 3 months), the cities will then organize all the data gathered and compare them against a similar timeframe under normal circumstances when the toll is in effect.

### Statistical Analysis

**2-sample pooled t-test for each city**\
The rationale is that first there are 2 distributions being studied:
1. The activity of each city under normal circumstances with tolls over a 3 month period
2. The activity of each city under the experiment with all tolls reimbursed over a 3 month period

Important to note is that the sample that is being studied here is the same in both distributions, as it is the same sample that underwent both the normal circumstance and the experimental circumstance.\
Using assumptions made above, consider using GDP as an analog for measurement of activity in a day. A data point in each distribution would be GDP generated within an hour. This way, the differences within each day can be visible.\
The pooled t-test will be performed on each city, as the goal was to see if the experiment caused a significant increase in activity in each city. The emphasis for Gotham is weekday daytime activity, and for Metropolis is weekday nighttime activity.

### Interpretation of Results

The hypotheses for both cities would be the each for the following:\
**H~0: The experimental mean GDP is not greater than the normal mean GDP**\
**H~1: The experimental mean GDP is greater than the normal mean GDP**

If the null hypothesis could not be rejected, then we can conclude the experiment to reimburse all toll fees did not significantly improve mean activity across both cities. Otherwise, if the null hypothesis can be rejected, we can say the experiment yielded significant improvement in activity across both cities.\
Even if the null hypothesis did get rejected, it might not mean however the experiment is a success, as in order for that to be, the increase in activity and GDP must be greater than the amount generated by the toll bridge under normal circumstances.
