```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# load the data
pd_users = pd.read_csv("takehome_users.csv", encoding='ISO-8859-1')
pd_engagement = pd.read_csv("takehome_user_engagement.csv")
```

```python
pd_users.head()
```

```python
pd_engagement.head()
```

```python
print("users shape and engagement shape:", pd_users.shape, pd_engagement.shape)
```

```python
pd_users.drop(columns=['name', 'email'], inplace=True)
pd_engagement.drop(columns=['visited'], inplace=True)
```

```python
pd_users.head()
```

```python
pd_engagement.head(30)
```

```python
test = pd_engagement.groupby(by='user_id').count()
test.head(30)
```

```python
# Let's use the engagement dataframe to calculate our adopted user condition
potential_users = test[test['time_stamp'] > 2]
potential_users
```

```python
returning_users = np.array(potential_users.index)
print(returning_users)
print(len(returning_users))
```

```python
# Apply the returning_users array as a mask to filter out all users who did not return at least 3 times to be considered an adopted user
pd_engagement = pd_engagement[pd_engagement['user_id'].isin(returning_users)]
pd_engagement.head(30)
```

```python
pd_engagement.shape
```

```python
pd_users.dtypes
```

```python
pd_users['date'] = pd.to_datetime(pd_users['last_session_creation_time'],unit='s')
```

```python
pd_users.head(30)
```

```python
pd_users.sort_values('date', ascending=False)
```

```python
# code for inverting the operation to convert to unix epoch
# pd.to_datetime(['2019-01-15 13:30:00']).map(pd.Timestamp.timestamp)
pd_engagement['unix_time'] = pd.to_datetime(pd_engagement['time_stamp']).map(pd.Timestamp.timestamp)
```

```python
pd_engagement.head(30)
```

```python
# 7-day period = 60*60*24*7 = 604800 seconds
adopted_user_mark = 60*60*24*7
print(adopted_user_mark)
```

```python
pd_engagement['third_unix_time'] = pd_engagement['unix_time'].shift(-2)
pd_engagement.head(30)
```

```python
pd_engagement['unix_diff'] = pd_engagement['third_unix_time'] - pd_engagement['unix_time']
pd_engagement.head(30)
```

```python
pd_engagement['adpoted_user'] = np.where(pd_engagement['unix_diff'] < adopted_user_mark, 1, 0)
pd_engagement.head(30)
```

```python
#adopter_user determines if the person is an adopter user
#unix diff provides the time difference needed to calculate if the adopted user threshold is met
#now need to remove the last 2 rows of every user
pd_engagement['user_id_shifted'] = pd_engagement['user_id'].shift(-2)
```

```python
pd_engagement['user_diff'] = pd_engagement['user_id'] - pd_engagement['user_id_shifted']
```

```python
df_adapted_user = pd_engagement.loc[pd_engagement['user_diff'] == 0]
```

```python
df_adapted_user.head(30)
```

```python
df_adapted_user = df_adapted_user.drop(df_adapted_user[df_adapted_user.adpoted_user != 1].index)
```

```python
df_adopted_user = df_adapted_user.drop_duplicates(subset=['user_id'])
```

```python
adopted_users = df_adopted_user.groupby('user_id').count()
```

```python
adopted_users
```

```python
# Create an array with only the users considered adopted
adopted_users = np.array(adopted_users.index)
```

```python
# Apply the adopted_users array as a mask to filter out all users who did not return at least 3 times to be considered an adopted user
pd_users['adopted_user'] = np.where(pd_users['object_id'].isin(adopted_users), 1, 0)
```

```python
# Now lets make all the columns of the users numeric to create a correlation heatmap to identify potential predictors
mapping = {
        'PERSONAL_PROJECTS': 1,
        'GUEST_INVITE': 2,
        'ORG_INVITE': 3,
        'SIGNUP': 4,
        'SIGNUP_GOOGLE_AUTH': 5
}
pd_users['creation_source'] = pd_users['creation_source'].map(mapping)
```

```python
pd_users['creation_time_unix'] = pd.to_datetime(pd_users['creation_time']).map(pd.Timestamp.timestamp)
```

```python
pd_users.dtypes
```

```python
pd_users_corr = pd_users.drop(columns=['creation_time', 'date'])
```

```python
f, ax = plt.subplots(figsize=(12,10))
sns.heatmap(pd_users_corr.corr(), annot=True, cmap="coolwarm",fmt='.2f');
f.suptitle("Correlation Heatmap")
```

### Results

The key target is adopted_user. Based on the heatmap above, it seems the best predictor for a user being labeled as adopted is their last_session_creation_time. Every other data point seems to have a much less effect in predicting them being a long term user. From this, we can guess that other good indicators of predicting adopted_user could be user_session_creation_frequency that calculates how often they create a session from when they created their first session to the current time/latest time in the entries. Other similar metrics could be one calculating frequency during each user's first month, as well as most recent month.
