## Political Polarization in US Presidential + Senatorial Elections

### Hypothesis: Polarization has increased in the most recent election cycles via states voting for same party in both races

If more and more states are not vote-splitting, then more and more are voting for the same party down the ballot, i.e. president and senate candidate from the same party.

```python
#Load the datasets
presidential_df = pd.read_csv("1976-2020-president.csv")

senate_df = pd.read_csv("1976-2020-senate.csv", encoding='ISO-8859-1')
```

```python
presidential_df.head()
```

```python
senate_df.head()
```

Begin with some data wrangling by dropping columns containing non-relavent information, creating a new column indicating the winner in each race, and reshaping the data so that the winners in each race in each state can be compared properly.

```python
drop_columns_pres = ['state_po', 'state_fips', 'state_cen', 'state_ic', 'party_detailed', 'writein', 'version', 'notes']
drop_columns_sen = ['state_po', 'state_fips', 'state_cen', 'state_ic', 'district', 'stage', 'special', 'party_detailed', 'writein', 'mode', 'unofficial', 'version']
```

```python
presidential_df = presidential_df.drop(columns=drop_columns_pres)
```

```python
senate_df = senate_df.drop(columns=drop_columns_sen)
```

```python
presidential_df.head()
```

```python
senate_df.head()
```

```python
#Fortunately, it seems for each state, the candidate with the most votes was ordered first
state_names = ['ALABAMA', 'ALASKA', 'ARIZONA', 'ARKANSAS', 'CALIFORNIA', 'COLORADO', 'CONNECTICUT', 'DELAWARE', 'FLORIDA', 'GEORGIA',
               'HAWAII', 'IDAHO', 'ILLINOIS', 'INDIANA', 'IOWA', 'KANSAS', 'KENTUCKY', 'LOUISIANA', 'MAINE', 'MARYLAND',
               'MASSACHUSETTS', 'MICHIGAN', 'MINNESOTA', 'MISSISSIPPI', 'MISSOURI', 'MONTANA', 'NEBRASKA', 'NEVADA', 'NEW HAMPSHIRE', 'NEW JERSEY',
              'NEW MEXICO', 'NEW YORK', 'NORTH CAROLINA', 'NORTH DAKOTA', 'OHIO', 'OKLAHOMA', 'OREGON', 'PENNSYLVANIA', 'RHODE ISLAND', 'SOUTH CAROLINA',
              'SOUTH DAKOTA', 'TENNESSEE', 'TEXAS', 'UTAH', 'VERMONT', 'VIRGINIA', 'WASHINGTON', 'WEST VIRGINIA', 'WISCONSIN', 'WYOMING']
```

```python
years = [1976, 1980, 1984, 1988, 1992, 1996, 2000, 2004, 2008, 2012, 2016, 2020]
```

```python
# Every year: 50 states hold elections
# Each state: there is a winner
# sort the values by: [year, state, candidatevotes]
sorted_pres_df = presidential_df.sort_values(['year', 'state', 'candidatevotes'], ascending=[True, True, False])
sorted_pres_df.head(30)
```

```python
sorted_pres_df['winner'] = (sorted_pres_df['candidatevotes'] == sorted_pres_df.groupby(['year', 'state'])['candidatevotes'].transform('max'))
```

```python
sorted_pres_df.head(30)
```

```python
#Now apply this to the real dataframes
presidential_df.sort_values(['year', 'state', 'candidatevotes'], ascending=[True, True, False], inplace=True)
senate_df.sort_values(['year', 'state', 'candidatevotes'], ascending=[True, True, False], inplace=True)
```

```python
presidential_df['winner'] = (presidential_df['candidatevotes'] == presidential_df.groupby(['year', 'state'])['candidatevotes'].transform('max'))
senate_df['winner'] = (senate_df['candidatevotes'] == senate_df.groupby(['year', 'state'])['candidatevotes'].transform('max'))
```

```python
presidential_df.head(30)
```

```python
senate_df.head(30)
```

```python
# Now remove all the candidates who did not win each race, leaving only the winners, and thus making the dataframes much cleaner
```

```python
presidential_winner = presidential_df[presidential_df['winner'] == True]
senate_winner = senate_df[senate_df['winner'] == True]
```

```python
presidential_winner.head(30)
```

```python
senate_winner.head(30)
```

```python
# Combine the 2 dataframes into one along the year and the state
```

```python
combined_df = presidential_winner.merge(
    senate_winner,
    on=["year", "state"],
    suffixes=("_pres", "_senate")
)
combined_df.head(30)
```

```python
combined_df.shape
```

```python
combined_df.tail(30)
```

```python
#Now lets see how many of these elections, the same party won all races in each state, separated by year.
parties = ['REPUBLICAN', 'DEMOCRAT', 'OTHER']
party_sweep = {}
total_elections = {}
year_ind = 0
year_count = 0
year_total = 0
for row in combined_df.itertuples():

    #Check to see if the election year has changed
    if row.year != years[year_ind]:
        #New election year means update dict with previous election data and advance the year and reset the counter
        party_sweep.update({years[year_ind]: year_count})
        total_elections.update({years[year_ind]: year_total})
        year_ind += 1
        year_count = 0
        year_total = 0
    #Increment count if one party won both races in the same state
    if row.party_simplified_pres == row.party_simplified_senate:
        year_count += 1
    #Counts total number of elections in year with both presidential and senate
    year_total += 1

#Update the final year now that the for loop has finished
party_sweep.update({years[year_ind]: year_count})
total_elections.update({years[year_ind]: year_total})
```

```python
party_sweep
```

```python
total_elections
```

```python
#Let's now see how many times it was the democrats who swept a state and how many times the republicans swept a state
dem_sweep = {}
gop_sweep = {}
year_ind = 0
dem_count = 0
gop_count = 0
for row in combined_df.itertuples():

    #Check to see if the election year has changed
    if row.year != years[year_ind]:
        #New election year means update dict with previous election data and advance the year and reset the counter
        gop_sweep.update({years[year_ind]: gop_count})
        dem_sweep.update({years[year_ind]: dem_count})
        year_ind += 1
        gop_count = 0
        dem_count = 0
    
    if row.party_simplified_pres == row.party_simplified_senate and row.party_simplified_pres == parties[0]:
        gop_count += 1
    elif row.party_simplified_pres == row.party_simplified_senate and row.party_simplified_pres == parties[1]:
        dem_count += 1

#Update the final year now that the for loop has finished
dem_sweep.update({years[year_ind]: dem_count})
gop_sweep.update({years[year_ind]: gop_count})
```

```python
dem_sweep
```

```python
gop_sweep
```

```python
#A list showing which party won the presidency for each year from 1976-2020
president_elect = ["DEMOCRAT", "REPUBLICAN", "REPUBLICAN", "REPUBLICAN", "DEMOCRAT", "DEMOCRAT",
                   "REPUBLICAN", "REPUBLICAN", "DEMOCRAT", "DEMOCRAT", "REPUBLICAN", "DEMOCRAT"]
```

### Now to the plots

```python
party_sweeps = list(party_sweep.values())
```

```python
total_count = list(total_elections.values())
```

```python
df_1 = pd.DataFrame({"election year": years,
                    "party sweeps": party_sweeps,
                  "all elections": total_count,})
df_1
```

```python
sns.lineplot(x='election year', y='value', hue='variable',
             data=pd.melt(df_1, ["election year"]),
            palette=['purple', 'black'])

#plt.xticks(ticks=np.arange(0, 91, 5))
plt.title("Party Sweeps in States, Presidential + Senate")
plt.xlabel("Election Years")
plt.ylabel("Election Count")
plt.show()
```

```python
democratic_sweeps = list(dem_sweep.values())
republican_sweeps = list(gop_sweep.values())
```

```python
df_2 = pd.DataFrame({"election year": years,
                    "democratic sweeps": democratic_sweeps,
                  "republican sweeps": republican_sweeps,
                    "president elect": president_elect})
df_2
```

```python
sns.lineplot(x='election year', y='value', hue='variable',
             data=pd.melt(df_2.drop(columns='president elect'), ["election year"]),
            palette=['blue', 'red'])

plt.title("Party Sweeps in States by Party, Presidential + Senate")
plt.xlabel("Election Years")
plt.ylabel("Election Count")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(df_2['election year'], df_2['democratic sweeps'], label='Democratic', color='blue')
ax.bar(df_2['election year'], df_2['republican sweeps'], label='Republican', bottom=df_2['democratic sweeps'], color='red')

ax.set_title("Party Sweeps in States by Party, Presidential + Senate")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(df_2['election year'], df_2['democratic sweeps'], label='Democratic', color='blue')
ax.bar(df_2['election year'], df_2['republican sweeps'], label='Republican', bottom=df_2['democratic sweeps'], color='red')

party_colors = {"REPUBLICAN": "mistyrose", "DEMOCRAT": "lightblue"}

for i, year in enumerate(years):
    winner = president_elect[i]
    color = party_colors[winner]
    
    # Define x-span: from this year - 2 to this year + 2 (to cover around the bar)
    # Adjust the +/- to control how much width you want shaded
    ax.axvspan(year - 2, year + 2, facecolor=color, alpha=0.5, zorder=-1)

ax.set_xlim(min(years) - 3, max(years) + 3)
ax.set_title("Party Sweeps in States by Party, Presidential + Senate")
ax.set_xlabel("Election Years")
ax.set_ylabel("Election Count")
ax.legend()
plt.show()
```

### Conclusion

It appears that same party sweeps have dramatically increased in the recent election cycles, peaking in 2016 when all states voted along party lines in both races. When factoring in the winner of the presidential election of each cycle, there appears to be some correlation where the party that won the presidency saw more races where the state voted for them on both the presidential level and the senate race. In other words, whichever party won the presidency tends to win more state sweeps than the losing party. Notably, The years of 1996 and 2000 though did tilt the other way.
